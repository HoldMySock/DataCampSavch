```python
import pandas as pd
import numpy as np

try:
    energy = pd.read_excel('Energy Indicators.xls', skiprows=17, skipfooter=38, usecols=[2, 3, 4, 5], names=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'], na_values='...')
    print("Successfully read Energy Indicators.xls")
except FileNotFoundError:
    print("Error file not found")

energy['Energy Supply'] = energy['Energy Supply'] * 1000000

energy.head(5)
```

```python
energy['Country'] = energy['Country'].str.replace(r'\d+', '', regex=True).str.replace(r'\(.*?\)', '', regex=True).str.strip()

energy['Country'] = energy['Country'].replace({
    'Republic of Korea': 'South Korea',
    'United States of America': 'United States',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'China, Hong Kong Special Administrative Region': 'Hong Kong'
})

energy[energy['Country'] == 'South Korea']
```

```python
try:
    GDP = pd.read_csv('API_NY.GDP.MKTP.CD_DS2_en_csv_v2_133326.csv', skiprows=4, skipfooter=1, delimiter=",", engine='python')
    print("Successfully read API_NY.GDP.MKTP.CD_DS2_en_csv_v2_133326.csv")
except FileNotFoundError:
    print("Error file not found")

GDP.head(3)
```

```python
GDP['Country Name'] = GDP['Country Name'].replace({
    'Korea. Rep': 'South Korea',
    'Iran, Islamic Rep.': 'Iran',
    'Hong Kong SAR, China': 'Hong Kong',
})

GDP[GDP['Country Name'] == 'Iran']
```

```python
try:
    ScimEn = pd.read_excel('scimagojr country rank 1996-2024.xlsx')
    print("Successfully read from scimagojr country rank 1996-2024.xlsx")
except FileNotFoundError:
    print("Error file not found")

ScimEn.head(3)
```

```python
def answer_one():
    gdp_years = [str(y) for y in range(2006, 2016)]
    GDP_filtered = GDP[['Country Name'] + gdp_years].rename(columns={'Country Name': 'Country'})

    top15_ScimEn = ScimEn[ScimEn['Rank'] <= 15]

    JointDF = top15_ScimEn.merge(energy, on='Country', how='inner') \
                          .merge(GDP_filtered, on='Country', how='inner')

    JointDF = JointDF.set_index('Country')
    cols = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'] + gdp_years
    return JointDF[cols]

answer_one()
```

```python
def answer_two():
    df = answer_one()
    gdp_years = [str(y) for y in range(2006, 2016)]
    avgGDP = df[gdp_years].mean(axis=1, skipna=True).sort_values(ascending=False)
    avgGDP.name = 'avgGDP'
    return avgGDP

answer_two()
```

```python
def answer_four():
    df = answer_one().copy()
    df['Ratio'] = df['Self-citations'] / df['Citations']
    country = df['Ratio'].idxmax()
    return f"{country} {df.loc[country, 'Ratio']:.12f}"

answer_four()
```

```python
def answer_five():
    df = answer_one().copy()
    df['PopEst'] = df['Energy Supply'] / df['Energy Supply per Capita']
    return df['PopEst'].nlargest(3).index[2]

answer_five()
```

```python
def answer_six():
    df = answer_one().copy()
    df['PopEst'] = df['Energy Supply'] / df['Energy Supply per Capita']
    df['Citable docs per Capita'] = df['Citable documents'] / df['PopEst']
    return float(df['Citable docs per Capita'].corr(df['Energy Supply per Capita']))

answer_six()
```

```python
ContinentDict = {'China':'Asia', 'United States':'North America', 'Japan':'Asia',
                 'United Kingdom':'Europe', 'Russian Federation':'Europe', 'Canada':'North America',
                 'Germany':'Europe', 'India':'Asia', 'France':'Europe', 'South Korea':'Asia',
                 'Italy':'Europe', 'Spain':'Europe', 'Iran':'Asia', 'Australia':'Australia',
                 'Brazil':'South America'}

def answer_seven():
    df = answer_one().copy()
    df['PopEst'] = df['Energy Supply'] / df['Energy Supply per Capita']
    df['Continent'] = pd.Series(ContinentDict)
    return df.groupby('Continent')['PopEst'].agg(['size', 'sum', 'mean', 'std'])

answer_seven()
```
