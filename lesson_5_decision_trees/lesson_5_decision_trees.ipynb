```python
import os
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn import tree
import graphviz
from sklearn.model_selection import train_test_split

os.environ["PATH"] += os.pathsep + r'C:\Program Files (x86)\Graphviz2.38\bin'
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X, y, labels, features = cancer.data, cancer.target, cancer.target_names, cancer.feature_names
print ('labels:',labels)
print ('features:', features)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
#Decision Tree Classifier

max_depth = 3
clf = DecisionTreeClassifier(
    criterion = 'entropy',
    random_state = 0,
    max_depth = max_depth,
    min_samples_leaf = 10,
    min_samples_split = 20, #good results (small gap between train and test result), I used min split and min leaf as constraints, can even hide max_depth with these params
).fit(X_train, y_train)

print("train accuracy= {:.3%}".format(clf.score(X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score(X_test, y_test)))
graph_viz = tree.export_graphviz(clf, out_file=None, feature_names=features, class_names=labels, filled=True)
graph = graphviz.Source(graph_viz)
graph.view(cleanup=True)
```

```python
#Random Forest Classifier

clf = RandomForestClassifier(
    random_state = 0,
    n_estimators = 200,
    max_features = 'sqrt',
    max_depth = 3,  #not too big of a gap between train and test
).fit(X_train, y_train)

print("train accuracy= {:.3%}".format(clf.score(X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score(X_test, y_test)))
```

```python
#Gradient Boosting Classifier

clf = GradientBoostingClassifier(
    n_estimators = 100,
    learning_rate = 0.01,
    max_depth = 2, #best results, small gap between train/test, good test accuracy
).fit(X_train, y_train)

print("train accuracy= {:.3%}".format(clf.score(X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score(X_test, y_test)))
```

```python
#XGBoost Classifier

clf = XGBClassifier(
    eval_metric='logloss',
    n_estimators=100,
    max_depth=3, #either 2 or 3, 2 is train accuracy= 96.948% test accuracy= 95.105%, 3 is train accuracy= 98.826% test accuracy= 97.203%
    learning_rate=0.01,
    reg_lambda=1, 
)
clf.fit(X_train, y_train)

print("train accuracy= {:.3%}".format(clf.score(X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score(X_test, y_test)))
```
